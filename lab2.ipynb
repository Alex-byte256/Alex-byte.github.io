```python
import numpy as np
import time
import unittest

def getColAsList(matrixToManipulate, col):
    myList = []
    numOfRows = len(matrixToManipulate)
    for i in range(numOfRows):
        myList.append(matrixToManipulate[i][col])
    return myList
 
def getCell(matrixA, matrixB, r, c):
    matrixBCol = getColAsList(matrixB, c)
    lenOfList = len(matrixBCol)
    productList = [matrixA[r][i]*matrixBCol[i] for i in range(lenOfList)]
    return sum(productList)

def vectorOnVector(vectorA, vectorB, lenght):
    scalar = 0
    for i in range(lenght):
        scalar = scalar + vectorA[i]*vectorB[i]
    return scalar



matrixA = np.random.rand(4,30)
matrixB = np.random.rand(30,5)
matrixC = np.random.rand(200,200)
vectorA = np.random.rand(200)
vectorB = np.random.rand(200)




rowA = len(matrixA)
colA = len(matrixA[0])
rowB = len(matrixB)
colB = len(matrixB[0])
rowC = len(matrixC)

matrixOnVector = [0] * (rowC)
vectorOnMatrix = [0] * (rowC)

start_time1 = time.time() 
result = [[0 for p in range(colB)] for q in range(rowA)]
print('\nMatrix x Matrix: ')
for i in range(rowA):
  for j in range(colB):
        result[i][j] = getCell(matrixA, matrixB, i, j)
  print(result[i])
print("--- %s s TIME ---" % (time.time() - start_time1))

start_time2 = time.time() 
print('\nMatrix x Vector: ')
for i in range(rowC):
   matrixOnVector[i] = 0
for i in range(rowC):
   for j in range(rowC):
     matrixOnVector[i] += (matrixC[i][j] * vectorA[j])
print(matrixOnVector)
print("--- %s s TIME ---" % (time.time() - start_time2))

start_time3 = time.time() 
print('\nVector x Matrix: ')
for i in range(rowC):
   vectorOnMatrix[i] = 0
for i in range(rowC):
   for j in range(rowC):
     vectorOnMatrix[i] += (matrixC[j][i] * vectorA[j])
print(vectorOnMatrix)
print("--- %s s TIME ---" % (time.time() - start_time3))

start_time4 = time.time() 
print('\nVector x Vector: ')
print(vectorOnVector(vectorA, vectorB, rowC))
print("--- %s s TIME ---" % (time.time() - start_time4))

start_time5 = time.time() 
print('\nMatrix x Matrix with Numpy: ')
print(np.matmul(matrixA, matrixB))
print("--- %s s TIME ---" % (time.time() - start_time5))

start_time6 = time.time() 
print('\nMatrix x Vector with Numpy: ')
print(np.matmul(matrixC, vectorA))
print("--- %s s TIME ---" % (time.time() - start_time6))

start_time7 = time.time() 
print('\nVector x Matrix with Numpy: ')
print(np.matmul(vectorA, matrixC))
print("--- %s s TIME ---" % (time.time() - start_time7))

start_time8 = time.time() 
print('\nVector x Vector with NumPy: ')
print(np.matmul(vectorA, vectorB))
print("--- %s s  TIME ---" % (time.time() - start_time8))
```
